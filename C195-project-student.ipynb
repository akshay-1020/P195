# Task 1 - Find the diabetic and hight blood pressure patients across all age groups, and conclude the risk heart stroke is higher in which age group

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt 

#read the csv
df = pd.read_csv('heart_disease.csv')
df
```

```python
#Filter by diabetes(condition will be who has diabetes) and create new dataframe

dpatient = df.loc[df['diabetes'] == 1]
dpatient
```

```python
#On this new data frame perform group operation as per age and create new dataframe 

dpatientgroup = dpatient.groupby('age')['diabetes'].count().reset_index()
dpatientgroup
```

```python
#Filter by high_blood_pressure(condition will be who has high_blood_pressure) and create new dataframe

hpatient = df.loc[df['high_blood_pressure'] == 1]
hpatient
```

```python
#On this new data frame perform group operation as per age and create new dataframe 

hpatientgroup = hpatient.groupby('age')['high_blood_pressure'].count().reset_index()
hpatientgroup
```

```python
#plot the scatter graph to show which age group is more prone to diabetes

age = dpatientgroup['age']
diabetes = dpatientgroup['diabetes']

plt.scatter(age, diabetes, label = "A1C Per Age")

plt.title("Diabetes Levels Per Age")
plt.xlabel("Age")
plt.ylabel("A1C")
plt.legend()
```

Conclusion - Ages 50-60 Have the worst diabetes, and 80-90 are the best years with the least diabetes.
